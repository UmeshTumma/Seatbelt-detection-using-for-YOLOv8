Video Demo 

```python
from ultralytics import YOLO
import cv2

# Load YOLO model
model = YOLO(r"C:\Umesh college\Project file\Seatbelt_Detection_Using _Yolov8new\Seatbelt_Detection\Seatbelt_Detection\weights\best.pt")

# Open video file
video_path = r"C:\Umesh college\Project file\Seatbelt_Detection_Using _Yolov8new\Demo Image SD\test_Trim.mp4"
cap = cv2.VideoCapture(video_path)

# Check if video is opened
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Get FPS for smooth display
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Process video
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO inference and get processed frame
    frame = model(frame)[0].plot()

    # Display output
    cv2.imshow("YOLOv8 Helmet Detection", frame)

    # Adjust delay for smooth visualization
    if cv2.waitKey(int(1000 / fps)) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```

Live Demo 

```python
import cv2
from ultralytics import YOLO

model = YOLO(r"Seatbelt_Detection/Seatbelt_Detection/weights/best.pt")
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    exit("Error: Could not open webcam.")

while True:
    success, frame = cap.read()
    if not success:
        break

    results = model(frame)
    for result in results:
        for box in result.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, f"{model.names[int(box.cls[0])]}: {box.conf[0]:.2f}", 
                        (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    cv2.imshow("Helmet Detection", frame)
    if cv2.waitKey(1) == ord("z"):
        break

cap.release()
cv2.destroyAllWindows()
```
